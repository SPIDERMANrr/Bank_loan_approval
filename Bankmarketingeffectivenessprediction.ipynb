````python
#function to get mean of a feature in the column
def get_mean(df,feature,col):
  new_df=df[df[feature]==col]
  final_df=pd.DataFrame()
  final_df=new_df.describe().iloc[1]
  return final_df
````

````python
#function to get median of a feature in the column
def get_median(df,feature,col):
  new_df=df[df[feature]==col]
  final_df=pd.DataFrame()
  final_df=new_df.describe().iloc[5]
  return final_df
````

````python
#function to get counts of a feature in the column
def get_count(df,feature,col):
  new_df=df[df[feature]==col]
  final_df=pd.DataFrame()
  final_df=new_df.describe().iloc[0]
  return final_df
````

````python
# creating list of unique values
marital_list=list(df['marital'].unique())
education_list=list(df['education'].unique())
job_list=list(df['job'].unique())
default_list=list(df['default'].unique())
housing_list=list(df['housing'].unique())
month_list=list(df['month'].unique())
poutcome_list=list(df['poutcome'].unique())
````

````python
#function to find the mean of the customers who combindly fall under the categories of feature 1 and feature 2

def mean_dataframe(df,list1,list2,feature1,feature2):
  dict_L1={}
  dict_L2={}
  for x in list1:
    dict_L2={}
    temp_df=df[df[feature1]==x]
    for i in list2:
      dict_L2[i]=get_mean(temp_df,feature2,i)
      pd.concat(dict_L2, names=[feature2],axis=1)
    dict_L1[x]=dict_L2
  temp_list=[]
  for x in list1:
    temp_list.append(pd.concat(dict_L1[x],axis=1))
  result = pd.concat(temp_list,axis=1,keys=list1)
  result.columns=['_'.join(col) for col in result.columns.values]
  result=result.T
  result.index.name='{} and {}'.format(feature1,feature2)
  return(result)
````

````python
#function to find the median of the customers who combindly fall under the categories of feature 1 and feature 2

def median_dataframe(df,list1,list2,feature1,feature2):
  dict_L1={}
  dict_L2={}
  for x in list1:
    temp_df=df[df[feature1]==x]
    for i in list2:
      dict_L2[i]=get_median(temp_df,feature2,i)
      pd.concat(dict_L2, names=[feature2],axis=1)
    dict_L1[x]=dict_L2
  temp_list=[]
  for x in list1:
    temp_list.append(pd.concat(dict_L1[x],axis=1))
  result = pd.concat(temp_list,axis=1,keys=list1)
  result.columns=['_'.join(col) for col in result.columns.values]
  result=result.T
  result.index.name='{} and {}'.format(feature1,feature2)
  return(result)
````

````python
#function to find the count of the customers who combindly fall under the categories of feature 1 and feature 2

def count_dataframe(df,list1,list2,feature1,feature2):
  dict_L1={}
  dict_L2={}
  for x in list1:
    dict_L2={}
    temp_df=df[df[feature1]==x]
    for i in list2:
      dict_L2[i]=get_count(temp_df,feature2,i)
      pd.concat(dict_L2, names=[feature2],axis=1)
    dict_L1[x]=dict_L2
  temp_list=[]
  for x in list1:
    temp_list.append(pd.concat(dict_L1[x],axis=1))
  result = pd.concat(temp_list,axis=1,keys=list1)
  result.columns=['_'.join(col) for col in result.columns.values]
  result=result.T
  result.index.name='{} and {}'.format(feature1,feature2)
  result=result['y']
  return(result)
````

````python
#function to created stacked bar charts of subscribed and not subscribed of a category
def myplotdf(df1,df2,ind):
  plotdata=pd.DataFrame({'subscribed':df1[ind],'not_subscribed':df2[ind]})
  plotdata=plotdata.reset_index(level=[1])
  plotdata['new']=plotdata.iloc[:,0].str.split('_').apply(lambda x: str(x[1]))
  plotdata['total']=plotdata['subscribed']+plotdata['not_subscribed']
  plotdata['percent_subscribed']=(plotdata['subscribed']/plotdata['total'])*100
  plotdata['percent_notsubscribed']=(plotdata['not_subscribed']/plotdata['total'])*100
  plotdata=plotdata.sort_values(by='new',ascending=True)
  # ploting graph
  n=len(df1.index)
  x_location=np.arange(n)
  bar_width=0.75
  fig, ax =plt.subplots(figsize=(80,50))
  #plot bar graph
  sbar=ax.bar(x_location,plotdata['subscribed'],bar_width,label='subscribed',color='cyan')
  nsbar=ax.bar(x_location,plotdata['not_subscribed'],bar_width,bottom=plotdata['subscribed'],label='not_subscribed',color='magenta')
  #plot a line for mean value of subscribed
  ax.axhline(plotdata['subscribed'].mean(),color='green',linewidth=2.0,label='mean of subscribed')
  ax.axhline(plotdata['total'].mean(),color='red',linewidth=2.0,label='mean of total')
  #add labels for the figure
  ax.set_xticks(x_location)
  ax.set_xticklabels(plotdata['level_1'],rotation=90)
  #insert legend
  ax.legend(loc='upper right')
  #insert data labels
  ax.bar_label(sbar,label_type='center')
  ax.bar_label(nsbar,label_type='center')
  ax.bar_label(nsbar,size=15)
  plt.show()
  return plotdata
````

## ***5. Hypothesis Testing***

### Based on the chart experiments,I have defined hypothetical statements from the dataset. In the question,I have performed hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Statement: Customers who are subscribing are atleast have a call duration of 600 seconds.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis: N >= 600

Alternate Hypothesis : N < 600

Test Type: Left Tailed Test


````python
#subscribed dataframe of duration
subscribed = df[df['y'] == 1]['duration']
````

````python
from scipy.stats import ttest_ind

# Subset the call duration data for each group to only include values above 600 seconds
subscribed_above_600 = subscribed[subscribed >= 800]
alpha=0.05
# Conduct t-test
t, p = ttest_ind(subscribed_above_600, [800]*len(subscribed_above_600))

# Print test statistic and p-value
print('t-test statistic:', t)
print('t-test p-value:', p)
````

````python
#Decision to accept or reject null hypothesis
if p < alpha:
    print("Reject null hypothesis. There is sufficient evidence to suggest that the mean call duration for subscribed customers is less than 600 seconds.")
else:
    print("Fail to reject null hypothesis. There is insufficient evidence to suggest that the mean call duration for subscribed customers is less than 600 seconds.")
````

##### Which statistical test have you done to obtain P-Value?

I have conducted a t-test to obtain the p-value. The t-test is a statistical test used to determine whether there is a significant difference between the means of two groups of data. In this case, we likely used a one-sample t-test to test whether the mean call duration for customers who are subscribed to the term deposit is significantly different from a given value of 600 seconds. The resulting t-test statistic and p-value provide us with information about the likelihood of observing the results we have obtained under the null hypothesis, which in this case is that the mean call duration for subscribed customers is equal to 600 seconds. A low p-value (e.g., less than 0.05) indicates that the observed difference between the sample mean and the null hypothesis value is statistically significant, and we can reject the null hypothesis in favor of the alternative hypothesis that the mean call duration is different from 600 seconds.

````python
# Handling Outliers & Outlier treatments
# To separate the symmetric distributed features and skew symmetric distributed features
symmetric_feature=[]
non_symmetric_feature=[]
for i in df.describe().columns:
  if abs(df[i].mean()-df[i].median())<0.2:
    symmetric_feature.append(i)
  else:
    non_symmetric_feature.append(i)

# Getting Symmetric Distributed Features
print("Symmetric Distributed Features : -",symmetric_feature)

# Getting Skew Symmetric Distributed Features
print("Skew Symmetric Distributed Features : -",non_symmetric_feature)

# Removing Customer Service Calls column from the list as it's an important factor
# which can't be treated as outliers here will is already leading to higher churn as we have seen furing analysis.
non_symmetric_feature.pop()
````

````python
# For Symmetric features defining upper and lower boundry
def outlier_treatment(df,feature):
  upper_boundary= df[feature].mean()+3*df[feature].std()
  lower_boundary= df[feature].mean()-3*df[feature].std()
  return upper_boundary,lower_boundary
````

````python
# Restricting the data to lower and upper boundry
for feature in symmetric_feature:
  df.loc[df[feature]<= outlier_treatment(df=df,feature=feature)[1], feature]=outlier_treatment(df=df,feature=feature)[1]
  df.loc[df[feature]>= outlier_treatment(df=df,feature=feature)[0], feature]=outlier_treatment(df=df,feature=feature)[0]
````

````python
# For Skew Symmetric features defining upper and lower boundry
#Outer Fence
def outlier_treatment_skew(df,feature):
  IQR= df[feature].quantile(0.75)- df[feature].quantile(0.25)
  lower_bridge =df[feature].quantile(0.25)-3*IQR
  upper_bridge =df[feature].quantile(0.25)+3*IQR
  return upper_bridge,lower_bridge
````

````python
# Restricting the data to lower and upper boundry
for feature in non_symmetric_feature:
  df.loc[df[feature]<= outlier_treatment_skew(df=df,feature=feature)[1], feature]=outlier_treatment_skew(df=df,feature=feature)[1]
  df.loc[df[feature]>= outlier_treatment_skew(df=df,feature=feature)[0], feature]=outlier_treatment_skew(df=df,feature=feature)[0]
````

##### Do you think that dimensionality reduction is needed? Explain Why?

As per my knowledge, for this dataset dimensionality reduction is not required.

Essentially where high dimensions are a problem or where it is a particular point in the algorithm to dimension reduction.

Hard rules are hard to state, other than “after you have tried it, did it improve matters”, which isn’t always the most useful guidance.

Instead, looking at why we might want to do this we can get a bit of insight. Admittedly some of the following might blur together a bit at the edges but the aim is to give a flavour.

1. Our data are too big. 4 million rows. 50,000 columns… is there a lot of redundancy there? Building a model on this could be very expensive. Even relatively simple dimension reduction techniques like PCA can capture almost all of the information in a fraction of the memory if there are strong relationships (that can be linearly approximated) in the data.

2. We are over-fitting. If you build a model with tens of thousands of degrees of freedom but don’t have a lot of examples you can easily overfit. Dimension reduction is one way of handling this, though often not the the best

3. We want to bring in external data. OK, this is a bit different but worth a note. In applications like word2vec we want to build a classifier using an embedding. We may want to classify some text into different categories but with only a limited number of examples. The complexity of free text is vast but a low dimension embedding is much smaller and will not overfit so badly in a classifier. Building a low dimensional embedding on external text, applying it to the text to be classified then building a classifier is using dimension reduction to bring in external data.

4. We suffer from the curse of dimesnionality. Consider something like a nearest neighbour search. As the number of dimensions gets large we see some unwanted behaviour, especially if we are looking at things like euclidean distances. Projecting your data to a lower dimensional space for nearest neighbour, clustering or outlier detection can be both more robust and more meaningful.

5. Some tools are all about this. Collaborative filtering through matrix factorisation is an example. Can we approximately describe behaviour as a linear combination of a smaller number of preferences/behaviours?
